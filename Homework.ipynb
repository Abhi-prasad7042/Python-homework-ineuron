## Question3

Write a function which will take input as a dict and give me an output as a list of all the valueseven in case of 2 level nesting it should work.

```python
def value_dict(d):
    li = []
    
    for i in d.values():
        if type(i)==dict:
            for j in i.values():
                li.append(j)
        else:
            li.append(i)
            
    return li
```

```python
d= {"ab":1, "b":2, "c": {"d":1,"k":12}, "f":{"s":12}}
```

```python
value_dict(d)
```

## Question8

Write a function which will be able to show your system configuration

```python
import wmi

def system_config():
    c = wmi.WMI()   
    my_system = c.Win32_ComputerSystem()[0]
    
    print(f"Manufacturer: {my_system.Manufacturer}")
    print(f"Model: {my_system. Model}")
    print(f"Name: {my_system.Name}")
    print(f"NumberOfProcessors: {my_system.NumberOfProcessors}")
    print(f"SystemType: {my_system.SystemType}")
    print(f"SystemFamily: {my_system.SystemFamily}")
```

```python
system_config()
```

Write a function which can move a file from one directory to another directory.

## Question18

Write a function which can help you to filter only word file from a directory.

```python
import os

def filter_word(path):
    l1 = os.listdir(path)
    li = []
    for i in l1:
        j = i.split('.')
        if len(j)>1 and j[1]=='doc':
            j = ".".join(j)
            li.append(j)
    
    return li
```
